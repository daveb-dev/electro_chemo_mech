```python
import pandas as pd
import os
import sys
import matplotlib
import matplotlib.pyplot as plt
homedir = os.getcwd()
```

```python
df1 = pd.read_csv("lco_circle_bc_no_coupling_out.csv",skiprows=[1])
df2 = pd.read_csv("lco_circle_bc_out_rho_1e-2.csv",skiprows=[1])
df3 = pd.read_csv("lco_circle_bc_out.csv",skiprows=[1])
```

```python
plt.plot(df1["time"],df1["ave_conc_inner"])
plt.plot(df2["time"],df2["ave_conc_inner"])
plt.plot(df3["time"],df3["ave_conc_inner"])

ave_conc_err1 = abs(df1["ave_conc_inner"]-df2["ave_conc_inner"])/(df1["ave_conc_inner"])
ave_conc_err2 = abs(df1["ave_conc_inner"]-df3["ave_conc_inner"])/(df1["ave_conc_inner"])
```

```python
plt.plot(df1["time"],ave_conc_err1*100)
plt.plot(df1["time"],ave_conc_err2*100)
```
