```python
import pandas as pd
import os
import sys
import matplotlib
import matplotlib.pyplot as plt
```

```python
pressure = ['0.0','10e-6','50e-6','100e-6']
pressure_dir = ['pres_' + x for x in pressure]
pressure_dir
```

```python
df_list = []
```

```python
# homedir = "/home/srinath/Documents/workspace_dev/electro_chemo_mech/test/diffusion_stress_one_way_coupling/chemo_mech1/lco/single/1C"
# os.chdir(homedir)
# os.getcwd()
```

```python
for i in pressure_dir:
    os.chdir(i)
    df_list.append(pd.read_csv('lco_circle_bc_out.csv',skiprows=[1,2]))
    os.chdir('../')
```

```python
for d,pr in zip(df_list, pressure):
    p = str(float(pr)*1e6) + "MPa"
    print pr, p
    plt.plot(d["time"]/7200.0, (d["ave_stress_22_top"] + float(pr))*1e12/1e5, label=p)
plt.legend()
plt.xlabel('Time/Time_period');
plt.ylabel(r'$\sigma_{11} / 10^5 $ (Pa)')
plt.savefig('time_vs_sigma11.png')
```

```python
for d,pr in zip(df_list, pressure):
    p = str(float(pr)*1e6) + "MPa"
    print pr, p
    plt.plot(d["ave_conc_inner"], (d["ave_stress_22_top"] + float(pr))*1e12/1e5, label=p)
plt.legend()
plt.xlabel('x (Li)')
plt.ylabel(r'$\sigma_{11} / 10^5 $ (Pa)')
plt.xlim([1.0,0.5])
plt.savefig('conc_vs_sigma11.png')
```
